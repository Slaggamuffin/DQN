```python
import gym
import csv
import copy
import random
import matplotlib.pyplot as plt
import torch
import torchvision.transforms as T
from torch.autograd import Variable
import numpy as np
import cv2
from IPython.display import clear_output
import time
from CPv3 import DQN
```

```python
model = torch.load('Results/DoubleDQN_ep300_lr000025')
env_name = 'CartPole-v1'
```

```python
class DQNModelVis(DQN):
    def __init__(self,env_name,model):
        super().__init__(env_name)
        self.model = model
    
    def render_model(self,episdoes):
        final = []
        for i in range(episdoes):
            state = self.reset()
            done = False
            count = 0
            
            while not done:
                q_values = self.predict('model',state)
                action = torch.argmax(q_values).item()

                next_state, reward, done, info = self.step(action)
                state = next_state
                
                if done:
                    final.append(count)
                    break
        
        self.env.close()
        


    
```

```python
agent = DQNModelVis(env_name,model)
agent.render_model(40)
```

```python
agent.env.close()
```
