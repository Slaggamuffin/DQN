```python
class A():
    def __init__(self, env_name):
        self.env_name = env_name
        self.episodes = 30
        print(self.env_name)
        
    def add(self,number):
        return self.episodes+number
    

class B(A):
    def __init__(self,model,env_name):
        super().__init__(env_name)
        self.model = model
        self.episodes = 40
        print(self.model,self.episodes)
        
    def wank(self,episodes):
        for i in range(episodes):
            count = self.add(10)

        print(count)

        
        
```

```python
a = A('env_name')
a = a.add(40)
```

```python
b = B('model','env_name')
b.add(40)
```
